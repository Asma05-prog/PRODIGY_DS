```python
import pandas as pd

df = pd.read_csv('titanic.csv')

print(df.head())
```

```python
print(df.info())
```

```python
print(df.isnull().sum())
```

```python
df['Age'] = df['Age'].fillna(df['Age'].median())

df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])

print(df.isnull().sum())
```

```python
print("Duplicates before:", df.duplicated().sum())

df.drop_duplicates(inplace=True)

print("Duplicates after:", df.duplicated().sum())
```

```python
print(df.describe())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

numeric_df = df.select_dtypes(include=['number'])

plt.figure(figsize=(10,6))

sns.heatmap(
    numeric_df.corr(),
    annot=True,
    cmap='coolwarm'
)

plt.title("Correlation Heatmap")

plt.show()
```

```python
df.hist(figsize=(12,10))

plt.show()
```

```python
sns.boxplot(x=df['Age'])

plt.title("Age Boxplot")

plt.show()
```

```python
sns.countplot(x='Survived', data=df)

plt.title("Survival Count")

plt.show()
```

Findings from EDA

1. The Age column contained missing values which were replaced using the median age.

2. The Embarked column contained missing values which were replaced using the most frequent value.

3. Most passengers were between 20 and 40 years old.

4. The dataset contains some age outliers as observed in the boxplot.

5. Survival is related to passenger characteristics such as age, class, and fare.

6. Correlation analysis helps identify relationships between numerical variables.
