```python
import pandas as pd

df = pd.read_csv('titanic.csv')

print(df.head())
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

plt.hist(df['Age'].dropna(), bins=20)

plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Number of Passengers")

plt.show()
```

```python
gender_count = df['Sex'].value_counts()

print(gender_count)

gender_count.plot(kind='bar')

plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Number of Passengers")

plt.show()
```

Histogram Interpretation

The histogram shows the distribution of passenger ages in the Titanic dataset.

Most passengers are between 20 and 40 years old.

There are fewer passengers in the higher age groups.

Bar Chart Interpretation

The bar chart shows the gender distribution of passengers.

Male passengers are more numerous than female passengers.

This indicates that the Titanic dataset contains more male passengers than female passengers.
