```python
import pandas as pd

df = pd.read_csv("US_Accidents_Sample.csv")

df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6,4))
sns.countplot(x='Severity', data=df)

plt.title("Accident Severity Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.countplot(y='Weather_Condition', data=df)

plt.title("Weather Condition Distribution")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.countplot(y='Road_Condition', data=df)

plt.title("Road Condition Distribution")

plt.show()
```

```python
plt.figure(figsize=(6,4))

sns.countplot(x='Sunrise_Sunset', data=df)

plt.title("Accidents During Day and Night")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.countplot(y='State', data=df)

plt.title("Accident Hotspots by State")

plt.show()
```
